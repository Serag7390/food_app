```python
import sqlite3 as sq
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATABASE_PATH = '/content/db.sqlite3'
```

```python
conn = sq.connect(DATABASE_PATH)
cur = conn.cursor()
```

```python
age_data = cur.execute('SELECT age from accounts_user').fetchall()
print(f'age_data: {age_data}')
```

```python
data = {

        'age': [i[0] for i in age_data]
}
sns.histplot(data = data, x='age')
```

```python
labels = 'young', 'adult', 'elder'
sizes = [
    len([1 for age in age_data if age[0] < 30])/ len(age_data),
    len([1 for age in age_data if 30 <= age[0] < 50])/ len(age_data),
    len([1 for age in age_data if age[0] >= 50])/ len(age_data)
]
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
INGREDIENTS = sorted(['egg','milk','butter','bread','cheese','rice', 'tomato', 'chiken', 'garlic', 'corn', 'cucumber', 'onion', 'lemon', 'cinnamon',
                      'fish', 'beard', 'beef', 'orange', 'mushroom', 'potato', 'carrot', 'olives', 'spinach', 'banana', 'peas', 'broccoli', 'apple', 'parsley', 'genger', 'pasta',
                      'chocolate', 'eggplant', 'cucumber'])
```

```python
def __extract_meal_ingredients(meals_ingredients):
    ingredients = []
    for meal_ingredients in meals_ingredients:
        for i, ingredient in enumerate(INGREDIENTS):
            if ingredient in meal_ingredients:
                ingredients.append(ingredient)
    return ingredients
```

```python
def get_data():
    meals_ingredients = cur.execute(f'''
    SELECT ingredients from pages_meal
    ''').fetchall()

    print(f'meals_ingredients: {len(meals_ingredients)}')

    meals_ingredients = [i[0] for i in meals_ingredients]
    data = __extract_meal_ingredients(meals_ingredients)
    print(f'data: {len(data)}')
    return data
```

```python
data = {
  'ingredients': sorted(get_data())
}
```

```python
plot = sns.histplot(data = data, x='ingredients')
plot.set_xticklabels(plot.get_xticklabels(), rotation=75)
```

```python
def get_data():
  orders = cur.execute(f'''
  SELECT quantity,ingredients from cart_order
  inner join pages_meal on cart_order.meal_id = pages_meal.id
  ''').fetchall()
  def flatten_orders(orders):
    flatten_orders = []
    for order in orders:
      flatten_orders.extend([order[1]] * order[0])
    return flatten_orders
  orders = flatten_orders(orders)
  data = __extract_meal_ingredients(orders)
  return data
```

```python
data = {
  'ingredients': sorted(get_data())
}
plot = sns.histplot(data = data, x='ingredients')
for item in plot.get_xticklabels():
  item.set_rotation(80)
```

```python
def get_data():
  orders = cur.execute(f'''
  SELECT quantity,ingredients, age  from cart_order
  inner join pages_meal on cart_order.meal_id = pages_meal.id
  inner join accounts_user on cart_order.user_id = accounts_user.id
  ''').fetchall()
  def flatten_orders(orders):
    flattened_orders = []
    for order in orders:
      flattened_orders.extend([(order[1],order[2])] * order[0])
    return flattened_orders
  orders = flatten_orders(orders)
  data = __extract_meal_ingredients([order[0] for order in orders])
  data = zip(data, [order[1] for order in orders])
  return list(data)
```

```python
age_ingredients_data = get_data()
print(age_ingredients_data[:10])

data = {
  'ingredients': [i[0] for i in age_ingredients_data],
  'age': [i[1] for i in age_ingredients_data]
}
```

```python
plt.scatter(data = data, y='age', x='ingredients')
plt.ylabel('age')
plt.xlabel('ingredients')
plt.xticks(rotation=80)
plt.show()
```
